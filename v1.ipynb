```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
sbn.set()
```

```python
data_dir = '../data/'
day_index1 = data_dir + '1day/index/2000_2009.csv'
day_index2 = data_dir + '1day/index/2010_2016.csv'
```

```python
indexes1 = pd.read_csv(day_index1,encoding='gbk', parse_dates=['date'])
indexes2 = pd.read_csv(day_index2,encoding='gbk', parse_dates=['date'])
```

```python
indexes = pd.concat([indexes1, indexes2])
```

```python
for column in ['open', 'high', 'low', 'close']:
    indexes[column] = indexes[column] / 10000.0
```

```python
idx300 = indexes[indexes['code']==300].copy()
```

```python
idx300.set_index(['date'], inplace=True)
```

```python
idx300.head()
```

```python
idx300['close'].shift(1)
```

```python
idx300n = pd.DataFrame()
for column in ['open', 'high', 'low', 'close']:
    pre_close = idx300['close'].shift(1)
    idx300n[column] = (idx300[column] - pre_close) / pre_close
```

```python
idx300n.dropna(inplace=True)
```

```python
idx300n = idx300n * 100.0
```

```python
idx300n['class'] = idx300n['close'].shift(-1)
```

```python
idx300n.dropna(inplace=True)
```

```python
idx300n.head()
```

```python
div_class = [-0.8, -0.5, -0.2, 0.2, 0.5, 0.8]
type_class = [-3, -2, -1, 0, 1, 2, 3]
def to_class(x):
    c = 0
    if x <= div_class[0]:
        c = type_class[0]
    elif div_class[0] <= x < div_class[1]:
        c = type_class[1]
    elif div_class[1] <= x < div_class[2]:
        c = type_class[2]
    elif div_class[2] <= x < div_class[3]:
        c = type_class[3]
    elif div_class[3] <= x < div_class[4]:
        c = type_class[4]
    elif div_class[4] <= x < div_class[5]:
        c = type_class[5]
    elif x >= div_class[5]:
        c = type_class[6]
    else:
        print("Error")
    return c
```

```python
idx300n['class']= idx300n['class'].map(to_class)
```

```python
values = idx300n.values
```

```python
train = values[:2500, :]
test = values[2500:2800, :]
```

```python
train_X, train_y = train[:, :-1], train[:, -1]
test_X, test_y = test[:, :-1], test[:, -1]
```

```python
train_X.shape
```

```python
test_X.shape
```

```python
train_X = train_X.reshape((-1, 50, train_X.shape[1]))
test_X = test_X.reshape((-1, 50, test_X.shape[1]))
```

```python
train_y.shape
```

```python
train_y = train_y.reshape((-1, 50, 7))
test_y = test_y.reshape((-1, 50, 7))
```

```python
print(train_X.shape, train_y.shape, test_X.shape, test_y.shape)
```

```python
from keras.utils import np_utils
```

```python
train_y = np_utils.to_categorical(train_y, num_classes=7)
test_y = np_utils.to_categorical(test_y, num_classes=7)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
```

```python
model = Sequential()
```

```python
units = 64
time_steps = 50
dim_input_vector = 4
no_classes = 7
# (timesteps, input_dim)
input_shape = (50, 4)
# batch_input_shape=()
```

```python
model.add(LSTM(units, activation='relu', dropout=0.5, return_sequences=True, input_shape=input_shape))
```

```python
model.add(LSTM(units, activation='relu', dropout=0.5, return_sequences=True))
```

```python
model.add(Dense(no_classes, activation='softmax'))
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(train_X, train_y, epochs=50, batch_size=30, validation_data=(test_X, test_y), verbose=2, shuffle=False)
```
